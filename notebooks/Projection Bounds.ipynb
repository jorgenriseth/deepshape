```python
import numpy as np
import matplotlib.pyplot as plt
```

## Comparison of Projection Approaches in the 1D-case 

```python
def f(x, ε=1e-1):
    return 1 / (1 -  np.minimum(0., x - ε))

def g(x, ε=1e-1):
    return (1 - ε) / (1 - np.minimum(ε, x))

# Noko. Er. På. Ferde.
ε = 1e-2

x = np.linspace(-2, 2, 401)
plt.figure(figsize=(12, 6))
plt.plot(x, f(x, ε), label='$f(x)$')
plt.plot(x, g(x, ε), label='$g(x)$')
plt.legend()
plt.ylabel("$k$")
plt.xlabel("$y_{min}$")
plt.axvline(ε, ls="--", c="k", lw=0.7)
plt.show()
```

# Projection Constants for Surfaces

```python
def Δ(k, a, b, ε):
    return a * k**2 + b * k + (1 - ε)
```

```python
ε = 1e-1
```

```python
k = np.linspace(-0.3, 1.3, 401)

plt.figure(figsize=(10, 6))
# plt.subplot(1, 2, 1)
plt.plot(k, Δ(k, 1., 2., ε), label='$b>0$')
plt.plot(k, Δ(k, 2., -1.5, ε), label='$ 0 > b > \sqrt{4a(1-ε)}$')
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.xticks([1], [1])
plt.yticks([1-ε], ['1-ε'])
plt.xlabel('$k$')
plt.ylabel('$Δ(k)$')
plt.ylim(-0.3, 1.5)
# plt.legend()


# plt.subplot(1, 2, 2)

plt.plot(k, Δ(k, 1., -4, ε), label='b>0')
plt.plot(k, Δ(k, 2.5, -3.1, ε), label='b>0')



ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.xticks([1], [1])
plt.yticks([1-ε], ['1-ε'])
plt.xlabel('$k$')
plt.ylabel('$Δ(k)$')


plt.ylim(-0.3, 1.5)
plt.savefig('projection_bounds.pdf')
plt.show()
```
